```python
import pandas as pd
import numpy as np

import requests
import json
```

On récupère le fichier JSON contenant nos valeurs de test

```python

chemin_fichier_json = "data_test100.json"

# Charger le fichier JSON
with open(chemin_fichier_json, "r") as fichier_json:
    data_json = json.load(fichier_json)
```

```python
print(data_json)
```

```python
# Définir l'URL de l'API
#url = "http://127.0.0.1:5000/"

url= "https://ludo-test-may-123.azurewebsites.net/"
#ludo-test-may-123.azurewebsites.net


# Effectuer la requête à l'API
response = requests.post(url, json=data_json)

# Vérifier si la requête a réussi (code de statut 200)   # tout s est bien passé
if response.status_code == 200:
    # Convertir la réponse JSON en un dictionnaire Python
    pred = response.json()
    # Afficher la réponse
    print(pred)
    print('\n')
    print("****La requête a été réalisé avec succès****")
else:
    # Afficher un message d'erreur si la requête a échoué
    print("****WARNING****!!La requête a échoué avec le code de statut:", response.status_code)





#content = json.loads(response.content.decode('utf-8'))

```
